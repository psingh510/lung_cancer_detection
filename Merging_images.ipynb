```python
import torchvision.transforms as transforms
from PIL import Image
import matplotlib.pyplot as plt
import os
import numpy as np
import cv2 as cv
import re
```

```python
def extract_position(input_name):
    # Use regex to find the 'Position' part
    match = re.search(r'Position \d+ - \d+_Z\d+_C\d+', input_name)
    if match:
        return match.group(0)
    else:
        return "Unknown_Position"
```

```python
folder1 = 'Images from Dellaire Lab/TIFF SAEC rep1 63x no As/dapi_new'
```

```python
folder2 = 'Images from Dellaire Lab/TIFF SAEC rep1 63x no As/pml_transformed'
```

```python
output_image_path = 'Images from Dellaire Lab/TIFF SAEC rep1 63x no As/merged_images'
```

```python
images1 = sorted(os.listdir(folder1))
images2 = sorted(os.listdir(folder2))
alpha = 1.5
beta = 0
for img1_name, img2_name in zip(images1,images2):
    img1_path = os.path.join(folder1, img1_name)
    img2_path = os.path.join(folder2, img2_name)
    print(img1_path)
    print(img2_path)
    image1 = cv.imread(img1_path)
    image2 = cv.imread(img2_path)
    merged_image = cv.addWeighted(image1, 0.5, image2, 0.5, 0)
    ## [basic-linear-transform-output]
    new_image = np.zeros(merged_image.shape, merged_image.dtype)
    ## [basic-linear-transform-output]

    for y in range(merged_image.shape[0]):
        for x in range(merged_image.shape[1]):
            for c in range(merged_image.shape[2]):
                new_image[y,x,c] = np.clip(alpha*merged_image[y,x,c] + beta, 0, 255)

    position = extract_position(img1_path)
    print(position)
    cv.imwrite(f'Images from Dellaire Lab/TIFF SAEC rep1 63x no As/merged_images/merged - {position}.jpg', new_image)
```
